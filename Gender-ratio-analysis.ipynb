```python
import pandas as pd
from plotnine import *
```

```python
df1 = pd.read_excel('2010-age-province-long.xlsx')
df2 = pd.read_excel('2010-age-province-rural-long.xlsx')
```

```python
df_pro_m = df1.groupby(['Province','Gender']).agg({'Number': 'sum'}).reset_index() \
        .pivot(index='Province', columns='Gender', values='Number')
df_pro_m['Ratio_marriage_g']= df_pro_m.M/ df_pro_m.F * 100
df_rural_m = df2.groupby(['Province','Gender']).agg({'Number': 'sum'}).reset_index() \
        .pivot(index='Province', columns='Gender', values='Number')
df_rural_m['Ratio_marriage_r']= df_rural_m.M/ df_rural_m.F * 100
```

```python
df_pro_m.to_csv("ratio_marriage.csv",index=None)
```
